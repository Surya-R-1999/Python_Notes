```python
import pandas as pd
```

```python
ratings_df = pd.read_csv("ratings.csv")
ratings_df.head()
```

```python
ratings_df.info()
```

```python
movies_df = pd.read_csv("movies.csv")
movies_df.head()
```

```python
movies_df.info()
```

```python
movies_df = movies_df.rename(mapper={"movieId":"movie_Id"},axis = 1)
```

```python
movies_df.head()
```

```python
result = pd.merge(ratings_df, movies_df, left_on='movieId',right_on="movie_Id", how='inner')
```

```python
result = result[["userId","movieId","title","timestamp"]]
```

```python
result.head()
```

```python
tags_df = pd.read_csv("tags.csv")
tags_df.head()
```

```python
final_result_inner = pd.merge(result,tags_df,on=["userId","movieId"],how="inner" , indicator=True,suffixes=('_left', '_right'))
final_result_inner.head()
```

```python
final_result_left = pd.merge(result,tags_df,on=["userId","movieId"],how="left", indicator=True,suffixes=('_left', '_right'))
final_result_left.head()
```

```python
final_result_right = pd.merge(result,tags_df,on=["userId","movieId"],how="right", indicator=True,suffixes=('_left', '_right'))
final_result_right.head()
```

```python
final_result_outer = pd.merge(result,tags_df,on=["userId","movieId"],how="outer", indicator=True,suffixes=('_left', '_right'))
final_result_outer.head()
```
