```python
import pandas as pd
```

having the Name as an index, Team as pivot column with salary as a value.

```python
data = pd.read_csv("EmployeesInfo.csv")
```

```python
data.info()
```

```python
data = data.drop_duplicates(subset = "First Name", keep = 'first')
```

```python
data.info()
```

```python
for i in data.columns:
    data[i].fillna(method='ffill',inplace = True)
```

```python
data.info()
```

```python
data['Gender'] = data['Gender'].astype('bool')
# data['Start Date'] = data['Start Date'].astype('DateTime')
# data['Last Login Time'] = data['Last Login Time'].dt.time
data['Senior Management'] = data['Senior Management'].astype('category')
```

```python
data.pivot(index = "Team", columns = "First Name", values = "Salary")
```
