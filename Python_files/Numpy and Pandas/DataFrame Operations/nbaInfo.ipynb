while updating a column in dataframe using list comprehension
-------------------------------------------------------------

You can update a column in a pandas dataframe using list comprehension as follows:

```python
df['column_name'] = [expression for item in df['column_name']]
```

Here, expression is the new value you want to assign to each item in the specified column. The list comprehension will iterate over each item in the specified column, apply the expression, and create a new list that replaces the original values in the column.

```python
import pandas as pd

# Create a sample dataframe
data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Update column 'B' using list comprehension
df['B'] = [val*2 for val in df['B']]

# Verify the result
print(df)
```

In this example, the values in column 'B' are multiplied by 2 using the list comprehension, and the result is assigned back to the same column.

```python
df["Name"].hasnans
```

To remove the record which consist atleast 1 null value
------------------------

```python
df.dropna(how = "any") # Even if any 1 column contains null value, the record is deleted 
```

```python
df.dropna(how = "all")  #if all column contains null values, then the record is deleted.
```

```python
df.dropna(subset = "College") # In this case, the specific column is mentioned, if the column College contains null, then the whole record is deleted.
```

```python
df.dropna(subset = ["College","Salary"]) 
```

To Fill the Nan Values in the dataset:
---------------------------------------

```python
df.fillna(0)
```

```python
df.fillna("ABC").info() #whenever we fill the null values with any special characters, then dtype will be changed as object.
```

```python
df.fillna(method = 'ffill') # filled with previous values
df.fillna(method = 'bfill') # filled with next values
```

```python
df['Salary'] = df['Salary'].fillna(0) #always fill the values with compatible datatype.
```

```python

df['College'] = df['College'].fillna("Texas")
```

```python
df["College"].hasnans
```
